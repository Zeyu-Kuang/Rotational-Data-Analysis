```python
# 06-07-2017
# This script is for plotting all the data in two data files
# located in '/home/ubuntu1/Documents/RotationData/test runs/' and '/home/ubuntu1/Documents/RotationData/run-9/'
# This is only for testing the data

__author__ = 'Zeyu'
from pandas import DataFrame
import matplotlib.pyplot as plt
```

```python
# test run (rms-fft)
folder1 = '/home/ubuntu1/Documents/RotationData/test runs/'  # data location
folder2 = '/home/ubuntu1/Documents/RotationData/run-9/'
dd0 = DataFrame.from_csv(folder1+'rms-fft.txt', sep='\t', header=None)  # read data into DataFrame
B0 = dd0.values  # change the data type to narray (numpy data structure)

# plot data inside
fig = plt.figure()
plt.plot(B0)
plt.show()
```

```python
# run-9 (data-file-K-0)
df0 = DataFrame.from_csv(folder2+'data-file-K-0', sep='\t', header=None, index_col=False)  # read data into DataFrame
A0 = df0.values  # change the data type to narray (numpy data structure)  

# print the narray data
print(A0)

"""
explaination of data structure
1 -- Time in normal time format
2 -- Sample number (1..N)
3 -- Measured voltage from Mixer (volts)
4 -- Measured voltage from PLL (volts)
5 -- Position from the encoder (in ticks) [there are 944000 counts per revelation][in run-8 and later the counter is reset at the homing mark, so 944000 is max value]
6 -- Time from the encoder (in micro-seconds) [1MHz clock is on board]
7 —- Ambient temperature in degree C (from run 10)
"""
```

```python
# plot each columns in the data
# figure for sample number
sn = A0[:,1]  # extract sample number
fig_sn = plt.figure()
plt.plot(sn)
plt.xlabel('array index', fontsize = 14)
plt.ylabel('sample number', fontsize = 14)
plt.show()
```

```python
# figure for measured voltage from Mixer 
vM = A0[:,2]
fig_vM = plt.figure()
plt.plot(sn,vM)
plt.xlabel('sample number', fontsize = 14)
plt.ylabel('voltage (volt)', fontsize = 14)
plt.title('voltage from Mixer (volt)', fontsize = 14)
plt.show()
```

```python
# figure for measured voltage from PLL
vP = A0[:,3]
fig_vP = plt.figure()
plt.plot(sn, vP)
plt.xlabel('sample number', fontsize = 14)
plt.ylabel('voltage (volt)', fontsize = 14)
plt.title('voltage from PLL (volt)', fontsize = 14)
plt.show()
```

```python
# figure for position from the encoder
position = A0[1:100,4]
fig_position = plt.figure()
plt.plot(sn[1:100], position)
plt.xlabel('sample number', fontsize = 14)
plt.ylabel('position (ticks)', fontsize = 14)
plt.title('Position from the encoder', fontsize = 14)
plt.show()
```

```python
# figure for time from the encoder 
time = A0[:,5]
fig_time = plt.figure()
plt.plot(sn, time)
plt.xlabel('sample number', fontsize = 14)
plt.ylabel('time (ms)', fontsize=14)
plt.title('Time form the encoder', fontsize=14)
plt.show()
```

```python
# figure for ambient temperature (not in the data sturcture)
temp = A0[:,6]
fig_temp = plt.figure()
plt.plot(sn, temp)
plt.xlabel('sample number', fontsize = 14)
plt.ylabel('Temperature (degree)', fontsize=14)
plt.title('Ambient temperature')
plt.show()
```
